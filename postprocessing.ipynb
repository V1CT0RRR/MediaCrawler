```python
import os
import json
import httpx
import asyncio
from pprint import pprint
from tqdm import tqdm
from rapidocr_onnxruntime import RapidOCR


semaphore = asyncio.Semaphore(value=4)

async def fetch_image(image_url: str, save_path: str, method: str="GET", timeout: int=10):
    async with semaphore:
        async with httpx.AsyncClient() as client:
            response = await client.request(
                method=method,
                url=image_url,
                timeout=timeout,
            )
            with open(save_path, "wb+") as file:
                file.write(response.content)
        print(f"[fetch_image] fetching from {image_url}")


async def save_images(note_json: dict):
    note_id = note_json.get("note_id")
    note_image_path = f"images/xhs/note_images/{note_id}"
    if not os.path.exists(note_image_path):
        os.mkdir(note_image_path)
    images = note_json.get("image_list", "").split(",")
    tasks = []
    for image_url in images:
        image_id = image_url.split("/")[-1]
        image_path = f"images/xhs/note_images/{note_id}/{image_id}.jpg"
        if os.path.exists(image_path):
            continue
        tasks.append({
            "image_url": image_url,
            "save_path": image_path,
        })

    print(f"[save_images] beginning saving images from {len(tasks)} tasks for {note_image_path}")

    async with semaphore:
        await asyncio.gather(*[
            fetch_image(**item) for item in tasks
        ])
    print(f"[save_images] done saving images from {len(tasks)} tasks for {note_image_path}")



async def ocr_image(ocr: RapidOCR, image_path: str, save_path: str):
    resp = ""
    result, _ = ocr(image_path)
    if result:
        ocr_result = [line[1] for line in result]
        resp += "\n".join(ocr_result)
    with open(save_path, "w+") as f:
        f.write(resp)
    return resp


async def ocr_images(note_images_path: str):
    ocr = RapidOCR()
    ocr_params = []
    for root, dirs, files in os.walk(note_images_path):
        for file in files:
            image_path = os.path.join(root, file)
            save_path = os.path.splitext(image_path)[0] + ".txt"
            if os.path.exists(save_path):
                continue
            ocr_params.append({
                "ocr": ocr,
                "image_path": image_path,
                "save_path": save_path,
            })

    ocr_results = []
    for item in tqdm(ocr_params, desc="[Postprocessing] OCR Images"):
        ocr_results.append(await ocr_image(**item))

    return ocr_results


async def process_images(json_file_path: str):
    notes = []
    with open(json_file_path, "r") as f:
        try:
            notes = json.loads(f.read())
        except json.JSONDecodeError:
            print(f"[Postprocessing] Invalid JSON from file: {json_file_path}")

    # await asyncio.gather(*[
    #     save_images(note_json=note_json) for note_json in notes
    # ])

    for note_json in notes:
        await save_images(note_json)

    print(f"[Postprocessing] Saved images from: {json_file_path}")

# await process_images("data/xhs/3_search_contents_2024-05-07.json")
```

```python
json_list = []
for root, dirs, files in os.walk("data"):
    for file in files:
        json_file_path = os.path.join(root, file)
        # await process_images(json_file_path)
        json_list.append(json_file_path)

for json_file_path in tqdm(json_list):
    await process_images(json_file_path)


print(f"[Postprocessing] Saved all images from: {json_list}")

note_images_path = "images/xhs/note_images"
if not os.path.exists(note_images_path):
    os.mkdir(note_images_path)

await ocr_images(note_images_path)
```

## Concat posts

```python
import os

json_list = []

for root, dirs, files in os.walk("data"):
    for file in files:
        json_file_path = os.path.join(root, file)
        # await process_images(json_file_path)
        json_list.append(json_file_path)

json_list
```

```python
import json
import pandas as pd

df_list = []

for json_item in json_list:
    with open(json_item, "r") as f:
        df_list.append(pd.DataFrame(json.loads(f.read())))
```

```python
df_posts = pd.concat(df_list).drop_duplicates(["note_id"]).reset_index(drop=True)

df_posts
```

```python
ocr_dicts = []

for root, dirs, files in os.walk("images"):
    note_id = os.path.split(root)[-1]
    image_contents = ""
    for file in files:
        if file.endswith(".txt"):
            with open(os.path.join(root, file), "r") as f:
                image_contents += f.read()
    ocr_dicts.append({
        "note_id": note_id,
        "ocr_result": image_contents,
    })

df_ocr_results = pd.DataFrame(ocr_dicts)

df_ocr_results
```

```python
# df_posts.join(df_ocr_results, on="note_id", how="left")
df_final = pd.merge(
    left=df_posts,
    right=df_ocr_results,
    how="left",
    left_on="note_id",
    right_on="note_id"
)
```
